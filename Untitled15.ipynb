***PCA***

```python
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X=df.drop(['label'],axis=1)
```

```python
X=X.values
#Scaling the values
x = scale(X)
```

```python
pca= PCA(n_components=20)
pca.fit(x)
```

```python
PCA(copy=True, iterated_power='auto', n_components=20, random_state=None,
    svd_solver='auto', tol=0.0, whiten=False)
```

```python
#The amount of variance that each PC explains
var= pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
var1
plt.plot(var1)
```

```python
#_________Eigen vector
pca_egvct = pca.components_
```

```python
#_________Eigen values
pca_egvl = pca.explained_variance_
```

```python
pca = PCA(n_components=15)
pca.fit(x)
X1=pca.fit_transform(x)
print(X1)
```

```python
x1=X1
y=df[['label']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x1,y,test_size=0.25)
```

***Model Building using Logistic Regression***

```python
from sklearn.linear_model import LogisticRegression
```

```python
model= LogisticRegression(solver='liblinear')
model.fit(x_train,y_train)
```

```python
print('Coefficient of model :', model.coef_)
print('Intercept of model',model.intercept_)
```

```python
#finding prediction of response variable
ypred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,ypred))
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
fpr, tpr, thresholds =roc_curve(y_test, ypred)
roc_auc = auc(fpr, tpr) 
print(roc_auc)
```

```python
#ROC Curve
plt.title('ROC Curve ')
plt.xlabel('False Positive Rate (1-Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.plot(fpr, tpr, label = 'AUC =' + str(roc_auc), lw=6)
plt.legend(loc=4) #Location of label
plt.show()
```
